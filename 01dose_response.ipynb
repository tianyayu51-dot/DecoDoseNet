```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import warnings
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

warnings.filterwarnings("ignore", category=RuntimeWarning)

# 从上传的CSV文件中加载数据
file_path = 'data/Hill/Single-dose-response.csv'
df = pd.read_csv(file_path)

# --- 统计独立药物和细胞系的数量 ---
num_drugs = df['drug_name'].nunique()
num_cell_lines = df['cell_line'].nunique()

print(f"独立药物数量: {num_drugs}")
print(f"独立细胞系数量: {num_cell_lines}")

# --- 定义四参数逻辑（4PL）模型函数 ---
def four_param_logistic(x, top, bottom, ec50, hill_slope):
    """用于剂量反应曲线拟合的四参数逻辑模型"""
    return bottom + (top - bottom) / (1 + (x / ec50)**hill_slope)

# --- 按药物和细胞系分组并进行曲线拟合 ---
fitted_params = []
grouped = df.groupby(['drug_name', 'cell_line'])

for (drug, cell_line), group in grouped:
    concentrations = group['Drug_concentration (µM)'].values
    responses = group['X/X0'].values
    
    # 初始化失败原因
    failure_reason = None
    
    # 检查数据是否足够
    if len(np.unique(concentrations)) < 3:
        failure_reason = f"浓度点不足({len(np.unique(concentrations))}个唯一浓度点)"
    elif len(np.unique(responses)) < 2:
        failure_reason = f"响应变化不足({len(np.unique(responses))}个唯一响应值)"
    
    # 如果数据不足，直接记录失败
    if failure_reason:
        print(f"拟合失败 {drug}-{cell_line}: {failure_reason}")
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.nan,
            'bottom': np.nan,
            'ec50': np.nan,
            'hill_slope': np.nan,
            'fit_success': False,
            'failure_reason': failure_reason
        })
        continue
    
    # 为参数提供初始猜测值
    try:
        top0 = max(responses)
        bottom0 = min(responses)
        ec50_0 = np.median(concentrations)
        hill0 = 1.0
        initial_guesses = [top0, bottom0, ec50_0, hill0]
        
        # 设置参数边界
        bounds = (
            [0, 0, 1e-6, 0.01],  # 下限
            [2.0, 2.0, 1e6, 10.0] # 上限
        )
        
        # 进行曲线拟合
        params, pcov = curve_fit(
            four_param_logistic, 
            concentrations, 
            responses, 
            p0=initial_guesses,
            bounds=bounds,
            maxfev=10000
        )
        
        # 计算R平方
        y_pred = four_param_logistic(concentrations, *params)
        r2 = r2_score(responses, y_pred)
        
        # 检查拟合参数是否合理
        top, bottom, ec50, hill_slope = params
        if not (bounds[0][0] <= top <= bounds[1][0]):
            failure_reason = f"top参数越界({top:.2f})"
        elif not (bounds[0][1] <= bottom <= bounds[1][1]):
            failure_reason = f"bottom参数越界({bottom:.2f})"
        elif not (bounds[0][2] <= ec50 <= bounds[1][2]):
            failure_reason = f"ec50参数越界({ec50:.2e})"
        elif not (bounds[0][3] <= hill_slope <= bounds[1][3]):
            failure_reason = f"hill_slope参数越界({hill_slope:.2f})"
        elif r2 < 0.5:
            failure_reason = f"拟合质量差(R²={r2:.2f})"
        
        if failure_reason:
            print(f"拟合失败 {drug}-{cell_line}: {failure_reason}")
            fitted_params.append({
                'drug_name': drug,
                'cell_line': cell_line,
                'top': top,
                'bottom': bottom,
                'ec50': ec50,
                'hill_slope': hill_slope,
                'r_squared': r2,
                'fit_success': False,
                'failure_reason': failure_reason
            })
        else:
            fitted_params.append({
                'drug_name': drug,
                'cell_line': cell_line,
                'top': top,
                'bottom': bottom,
                'ec50': ec50,
                'hill_slope': hill_slope,
                'r_squared': r2,
                'fit_success': True,
                'failure_reason': None
            })
            
    except RuntimeError as e:
        failure_reason = f"算法无法收敛: {str(e)}"
        print(f"拟合失败 {drug}-{cell_line}: {failure_reason}")
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.nan,
            'bottom': np.nan,
            'ec50': np.nan,
            'hill_slope': np.nan,
            'r_squared': np.nan,
            'fit_success': False,
            'failure_reason': failure_reason
        })
    except Exception as e:
        failure_reason = f"意外错误: {str(e)}"
        print(f"拟合失败 {drug}-{cell_line}: {failure_reason}")
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.nan,
            'bottom': np.nan,
            'ec50': np.nan,
            'hill_slope': np.nan,
            'r_squared': np.nan,
            'fit_success': False,
            'failure_reason': failure_reason
        })

# --- 创建并保存包含拟合参数的DataFrame ---
params_df = pd.DataFrame(fitted_params)
output_file = 'fitted_dose_response_parameters_with_failure_reasons.csv'
params_df.to_csv(output_file, index=False)

print(f"\n拟合参数已保存至 '{output_file}'")

# --- 输出拟合失败统计 ---
failed_fits = params_df[params_df['fit_success'] == False]
if not failed_fits.empty:
    print("\n===== 拟合失败统计 =====")
    print(f"总失败次数: {len(failed_fits)}/{len(params_df)}")
    print("\n失败原因分布:")
    print(failed_fits['failure_reason'].value_counts())
    
    # 输出前5个失败案例详情
    print("\n前5个失败案例详情:")
    print(failed_fits[['drug_name', 'cell_line', 'failure_reason']].head())
else:
    print("\n所有拟合均成功!")

# --- 可选：绘制第一个失败案例的诊断图 ---
if not failed_fits.empty:
    first_failure = failed_fits.iloc[0]
    fail_data = df[(df['drug_name'] == first_failure['drug_name']) & 
                 (df['cell_line'] == first_failure['cell_line'])]
    
    plt.figure(figsize=(10, 6))
    plt.scatter(fail_data['Drug_concentration (µM)'], fail_data['X/X0'])
    plt.xscale('log')
    plt.xlabel('药物浓度 (µM)')
    plt.ylabel('相对抑制率 (X/X0)')
    plt.title(f"失败案例: {first_failure['drug_name']}-{first_failure['cell_line']}\n"
              f"原因: {first_failure['failure_reason']}")
    plt.grid(True)
    plt.savefig('failed_fit_example.png')
    print("\n已保存失败案例诊断图 'failed_fit_example.png'")
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import warnings
import matplotlib.pyplot as plt

# 忽略scipy.optimize.curve_fit在拟合过程中可能产生的警告
warnings.filterwarnings("ignore", category=RuntimeWarning)

# 从上传的CSV文件中加载数据
try:
    file_path = 'data/Hill/Single-dose-response.csv'
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"错误：找不到文件 '{file_path}'。请确保文件路径正确。")
    exit()

# --- 统计独立药物和细胞系的数量 ---
num_drugs = df['drug_name'].nunique()
num_cell_lines = df['cell_line'].nunique()

print(f"独立药物数量: {num_drugs}")
print(f"独立细胞系数量: {num_cell_lines}")
print("-" * 30)

# --- 定义四参数逻辑（4PL）模型函数 ---
def four_param_logistic(x, top, bottom, ec50, hill_slope):
    """用于剂量反应曲线拟合的四参数逻辑模型"""
    return bottom + (top - bottom) / (1 + (x / ec50)**hill_slope)

# --- 按药物和细胞系分组并进行曲线拟合 ---
fitted_params = []
grouped = df.groupby(['drug_name', 'cell_line'])

for (drug, cell_line), group in grouped:
    concentrations = group['Drug_concentration (µM)'].values
    responses = group['X/X0'].values
    
    # --- 预检查 ---
    # 1. 数据点不足是硬伤，无法强制，必须跳过
    if len(concentrations) < 4:
        status = "失败：数据点不足"
        print(f"跳过 药物: {drug}, 细胞系: {cell_line} - {status} (需要至少4个点, 实际只有{len(concentrations)}个)")
        params_to_store = {'top': np.nan, 'bottom': np.nan, 'ec50': np.nan, 'hill_slope': np.nan}
    
    # 2. 检查浓度值是否有效
    elif np.any(concentrations <= 0):
        status = "失败：包含无效的药物浓度"
        print(f"跳过 药物: {drug}, 细胞系: {cell_line} - {status}")
        params_to_store = {'top': np.nan, 'bottom': np.nan, 'ec50': np.nan, 'hill_slope': np.nan}
        
    else:
        # --- 策略一：尝试使用边界约束进行拟合 ---
        # 为参数设置合理的生物学边界
        # top/bottom: 0-2 (允许轻微超出0-1范围)
        # ec50: 从一个很小的值到比最大浓度大100倍
        # hill_slope: 0.1-10 (覆盖大多数情况)
        bounds = ([0, 0, 1e-9, 0.1], [2, 2, np.max(concentrations) * 100, 10])
        initial_guesses = [max(responses), min(responses), np.median(concentrations), 1.0]

        try:
            params, _ = curve_fit(
                four_param_logistic, 
                concentrations, 
                responses, 
                p0=initial_guesses, 
                maxfev=10000,
                bounds=bounds # <-- 使用边界约束
            )
            params_to_store = {'top': params[0], 'bottom': params[1], 'ec50': params[2], 'hill_slope': params[3]}
            status = "成功"

        except RuntimeError:
            # --- 策略二：如果带边界的拟合仍失败，则进行规则化插补 ---
            print(f"注意: 药物: {drug}, 细胞系: {cell_line} - 拟合失败，执行强制规则。")
            status = "强制拟合 (曲线平坦或无法收敛)"
            
            # 应用规则
            mean_response = np.mean(responses)
            params_to_store = {
                'top': mean_response,
                'bottom': mean_response,
                'ec50': np.max(concentrations), # 关键规则：EC50设为最大测试浓度
                'hill_slope': 1.0 # 设为标准斜率
            }

    # 将本次结果存入列表
    result = {
        'drug_name': drug,
        'cell_line': cell_line,
        'fitting_status': status,
        **params_to_store
    }
    fitted_params.append(result)

# --- 创建并保存包含拟合参数和状态的DataFrame ---
params_df = pd.DataFrame(fitted_params)
params_df = params_df[['drug_name', 'cell_line', 'fitting_status', 'top', 'bottom', 'ec50', 'hill_slope']]

output_file = 'forced_fitted_dose_response_parameters.csv'
params_df.to_csv(output_file, index=False)

print(f"\n拟合参数及状态已保存至 '{output_file}'")
print("\n数据框的前5行 (可能包含强制拟合结果):")
print(params_df.head())

# --- 可选：绘制一个强制拟合的示例曲线以供观察 ---
forced_fit_example = params_df[params_df['fitting_status'].str.contains("强制拟合")]

if not forced_fit_example.empty:
    example_fit = forced_fit_example.iloc[0]
    example_drug = example_fit['drug_name']
    example_cell_line = example_fit['cell_line']
    
    example_data = df[(df['drug_name'] == example_drug) & (df['cell_line'] == example_cell_line)]
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False

    x_fit = np.logspace(np.log10(min(example_data['Drug_concentration (µM)'])), np.log10(max(example_data['Drug_concentration (µM)'])), 100)
    y_fit = four_param_logistic(x_fit, example_fit['top'], example_fit['bottom'], example_fit['ec50'], example_fit['hill_slope'])

    plt.figure(figsize=(8, 6))
    plt.scatter(example_data['Drug_concentration (µM)'], example_data['X/X0'], label='原始数据')
    plt.plot(x_fit, y_fit, 'r--', label='强制拟合曲线') # 使用虚线表示强制拟合
    plt.xscale('log')
    plt.xlabel('药物浓度 (µM) (对数坐标)')
    plt.ylabel('相对抑制率 (X/X0)')
    plt.title(f'强制拟合示例: 药物: {example_drug} | 细胞系: {example_cell_line}')
    plt.legend()
    plt.grid(True)
    plt.savefig('example_forced_fit_curve.png')
    print("\n已保存一个强制拟合的示例曲线图 'example_forced_fit_curve.png' 以供检查。")
else:
    print("\n本次运行没有发生强制拟合。")
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import warnings
import matplotlib.pyplot as plt

# 忽略scipy.optimize.curve_fit在拟合过程中可能产生的警告
warnings.filterwarnings("ignore", category=RuntimeWarning)

# --- 定义R平方计算函数 ---
def calculate_r_squared(y_true, y_pred):
    """
    计算拟合优度 R-squared (R²).
    """
    # 计算残差平方和
    ss_res = np.sum((y_true - y_pred) ** 2)
    # 计算总平方和
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    # 避免除以零的错误
    if ss_tot == 0:
        # 如果总变异为0（即所有y值都相同），则R²无意义
        return np.nan 
    r_squared = 1 - (ss_res / ss_tot)
    return r_squared

# 从上传的CSV文件中加载数据
try:
    file_path = 'data/Hill/Single-dose-response.csv'
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"错误：找不到文件 '{file_path}'。请确保文件路径正确。")
    exit()

# --- 统计独立药物和细胞系的数量 ---
num_drugs = df['drug_name'].nunique()
num_cell_lines = df['cell_line'].nunique()

print(f"独立药物数量: {num_drugs}")
print(f"独立细胞系数量: {num_cell_lines}")
print("-" * 30)

# --- 定义四参数逻辑（4PL）模型函数 ---
def four_param_logistic(x, top, bottom, ic50, hill_slope):
    """用于剂量反应曲线拟合的四参数逻辑模型"""
    return bottom + (top - bottom) / (1 + (x / ic50)**hill_slope)

# --- 按药物和细胞系分组并进行曲线拟合 ---
fitted_params = []
grouped = df.groupby(['drug_name', 'cell_line'])

for (drug, cell_line), group in grouped:
    concentrations = group['Drug_concentration (µM)'].values
    responses = group['X/X0'].values
    
    r_squared = np.nan # 默认为NaN
    
    # --- 预检查 ---
    if len(concentrations) < 4:
        status = "失败：数据点不足"
        params_to_store = {'top': np.nan, 'bottom': np.nan, 'ic50': np.nan, 'hill_slope': np.nan}
    elif np.any(concentrations <= 0):
        status = "失败：包含无效的药物浓度"
        params_to_store = {'top': np.nan, 'bottom': np.nan, 'ic50': np.nan, 'hill_slope': np.nan}
    else:
        # --- 策略一：尝试使用边界约束进行拟合 ---
        bounds = ([0, 0, 1e-9, 0.1], [2, 2, np.max(concentrations) * 100, 10])
        initial_guesses = [max(responses), min(responses), np.median(concentrations), 1.0]

        try:
            params, _ = curve_fit(
                four_param_logistic, 
                concentrations, 
                responses, 
                p0=initial_guesses, 
                maxfev=10000,
                bounds=bounds
            )
            params_to_store = {'top': params[0], 'bottom': params[1], 'ic50': params[2], 'hill_slope': params[3]}
            status = "成功"
            
            # --- 拟合成功后，计算R²进行评估 ---
            y_predicted = four_param_logistic(concentrations, *params)
            r_squared = calculate_r_squared(responses, y_predicted)

        except RuntimeError:
            # --- 策略二：如果带边界的拟合仍失败，则进行规则化插补 ---
            status = "强制拟合 (曲线平坦或无法收敛)"
            mean_response = np.mean(responses)
            params_to_store = {
                'top': mean_response,
                'bottom': mean_response,
                'ic50': np.max(concentrations),
                'hill_slope': 1.0
            }
            # R²对于强制拟合无意义，保持为NaN

    # 将本次结果存入列表
    result = {
        'drug_name': drug,
        'cell_line': cell_line,
        'fitting_status': status,
        **params_to_store,
        'r_squared': r_squared # 新增R²列
    }
    fitted_params.append(result)

# --- 创建并保存包含拟合参数和评估结果的DataFrame ---
params_df = pd.DataFrame(fitted_params)
# 调整列顺序
params_df = params_df[['drug_name', 'cell_line', 'fitting_status', 'r_squared', 'top', 'bottom', 'ic50', 'hill_slope']]

output_file = 'evaluated_fitted_parameters.csv'
params_df.to_csv(output_file, index=False)

print(f"\n拟合参数及评估结果已保存至 '{output_file}'")
print("\n数据框的前5行 (包含R²评估):")
print(params_df.head())

# --- 统计并输出低R²的比例 ---
print("-" * 30)
print("模型拟合质量评估:")

# 只对状态为“成功”的拟合进行评估
successful_fits_df = params_df[params_df['fitting_status'] == '成功'].copy()
# 剔除R²为NaN的特殊情况（尽管在成功拟合中不常见，但为确保代码健壮性）
successful_fits_df.dropna(subset=['r_squared'], inplace=True)

num_successful_fits = len(successful_fits_df)

if num_successful_fits > 0:
    # 计算R²小于0.7的数量
    poor_fits_count = (successful_fits_df['r_squared'] < 0.7).sum()
    # 计算比例
    proportion_poor_fits = poor_fits_count / num_successful_fits
    
    print(f"总计有 {num_successful_fits} 个成功拟合的曲线。")
    print(f"其中，有 {poor_fits_count} 个曲线的 R² 小于 0.7。")
    print(f"低质量拟合 (R² < 0.7) 的比例: {proportion_poor_fits:.2%}")
else:
    print("没有成功拟合的曲线可供评估。")

# （可选）可以绘制一个R²较低的例子，以供诊断
low_r2_example = successful_fits_df[successful_fits_df['r_squared'] < 0.7]
if not low_r2_example.empty:
    example_fit = low_r2_example.iloc[0]
    # ... 后续的绘图代码可以放在这里，用于展示拟合效果差的例子 ...
    print("\n提示: 数据中存在低 R² 的拟合，您可能需要检查这些数据点或模型适用性。")
```

### 强制拟合  
top = 响应均值

bottom = 最小响应

IC50 = 浓度中位数

Hill斜率 = 1.0

```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import warnings
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error

warnings.filterwarnings("ignore", category=RuntimeWarning)

# 读取数据
file_path = 'data/Hill/Single-dose-response.csv'
df = pd.read_csv(file_path)

# 统计数量
print(f"独立药物数量: {df['drug_name'].nunique()}")
print(f"独立细胞系数量: {df['cell_line'].nunique()}")

# 四参数逻辑模型
def four_param_logistic(x, top, bottom, ec50, hill_slope):
    return bottom + (top - bottom) / (1 + (x / ec50)**hill_slope)

# 计算评价指标的函数
def calculate_metrics(y_true, y_pred, n_params=4):
    n = len(y_true)
    r2 = r2_score(y_true, y_pred)
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - n_params - 1)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    
    # 计算AIC和BIC
    rss = np.sum((y_true - y_pred)**2)
    aic = n * np.log(rss/n) + 2 * n_params
    bic = n * np.log(rss/n) + n_params * np.log(n)
    
    return {
        'r_squared': r2,
        'adjusted_r_squared': adj_r2,
        'mse': mse,
        'rmse': rmse,
        'aic': aic,
        'bic': bic
    }

# 储存拟合参数
fitted_params = []
grouped = df.groupby(['drug_name', 'cell_line'])

for (drug, cell_line), group in grouped:
    concentrations = group['Drug_concentration (µM)'].values
    responses = group['X/X0'].values

    # 排除无效点
    if len(np.unique(concentrations)) < 3 or len(np.unique(responses)) < 2:
        print(f"数据不足以拟合: {drug} - {cell_line}")
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.nan,
            'bottom': np.nan,
            'ic50': np.nan,
            'hill_slope': np.nan,
            'r_squared': np.nan,
            'adjusted_r_squared': np.nan,
            'mse': np.nan,
            'rmse': np.nan,
            'aic': np.nan,
            'bic': np.nan,
            'fit_success': False
        })
        continue

    # 初始值猜测 + 边界设定
    top0 = max(responses)
    bottom0 = min(responses)
    ec50_0 = np.median(concentrations)
    hill0 = 1.0

    initial_guess = [top0, bottom0, ec50_0, hill0]
    bounds = (
        [0, 0, 1e-6, 0.01],          # 下界：top, bottom ≥ 0，EC50 > 0，slope > 0
        [2.0, 2.0, 1e6, 10.0]        # 上界
    )

    try:
        params, pcov = curve_fit(
            four_param_logistic,
            concentrations,
            responses,
            p0=initial_guess,
            bounds=bounds,
            maxfev=20000
        )
        
        # 计算预测值
        y_pred = four_param_logistic(concentrations, *params)
        
        # 计算评价指标
        metrics = calculate_metrics(responses, y_pred)
        
        # 获取参数的标准误差
        perr = np.sqrt(np.diag(pcov)) if np.isfinite(pcov).all() else [np.nan]*4
        
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': params[0],
            'top_stderr': perr[0],
            'bottom': params[1],
            'bottom_stderr': perr[1],
            'ic50': params[2],
            'ic50_stderr': perr[2],
            'hill_slope': params[3],
            'hill_slope_stderr': perr[3],
            'r_squared': metrics['r_squared'],
            'adjusted_r_squared': metrics['adjusted_r_squared'],
            'mse': metrics['mse'],
            'rmse': metrics['rmse'],
            'aic': metrics['aic'],
            'bic': metrics['bic'],
            'fit_success': True
        })
        
    except Exception as e:
        print(f"拟合失败 {drug}-{cell_line}，使用退化模型处理")
        # 回退：使用均值/线性等替代
        y_pred = np.full_like(responses, np.mean(responses))
        metrics = calculate_metrics(responses, y_pred)
        
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.mean(responses),
            'top_stderr': np.nan,
            'bottom': np.min(responses),
            'bottom_stderr': np.nan,
            'ic50': np.median(concentrations),
            'ic50_stderr': np.nan,
            'hill_slope': 1.0,
            'hill_slope_stderr': np.nan,
            'r_squared': metrics['r_squared'],
            'adjusted_r_squared': metrics['adjusted_r_squared'],
            'mse': metrics['mse'],
            'rmse': metrics['rmse'],
            'aic': metrics['aic'],
            'bic': metrics['bic'],
            'fit_success': False,
            'error_message': str(e)
        })

# 保存拟合参数
params_df = pd.DataFrame(fitted_params)
output_file = 'data/Hill/fitted_dose_response_parameters_with_metrics.csv'
params_df.to_csv(output_file, index=False)
print(f"\n拟合参数已保存至 '{output_file}'")
print(params_df.head())

# --- 绘图示例 ---
# 查找第一个成功的拟合结果
success_df = params_df[params_df['fit_success']]
if not success_df.empty:
    example_row = success_df.iloc[0]
    example_drug = example_row['drug_name']
    example_cell_line = example_row['cell_line']
    
    example_data = df[(df['drug_name'] == example_drug) & 
                     (df['cell_line'] == example_cell_line)]
    
    # 生成拟合曲线
    x_fit = np.logspace(np.log10(min(example_data['Drug_concentration (µM)'])), 
                        np.log10(max(example_data['Drug_concentration (µM)'])), 100)
    y_fit = four_param_logistic(
        x_fit,
        example_row['top'],
        example_row['bottom'],
        example_row['ic50'],
        example_row['hill_slope']
    )
    
    # 绘图设置
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    
    plt.figure(figsize=(10, 7))
    plt.scatter(example_data['Drug_concentration (µM)'], example_data['X/X0'], 
               label='原始数据')
    plt.plot(x_fit, y_fit, 'r-', label='4PL 拟合曲线')
    plt.xscale('log')
    plt.xlabel('药物浓度 (µM) (对数坐标)')
    plt.ylabel('相对抑制率 (X/X0)')
    
    # 在标题中添加关键指标
    plt.title(
        f'药物: {example_drug} | 细胞系: {example_cell_line}\n'
        f'IC50: {example_row["ic50"]:.3f} µM | '
        f'Hill斜率: {example_row["hill_slope"]:.2f}\n'
        f'R²: {example_row["r_squared"]:.3f} | '
        f'RMSE: {example_row["rmse"]:.3f}'
    )
    
    plt.legend()
    plt.grid(True)
    plt.savefig('example_dose_response_curve_with_metrics.png')
    print("\n已保存示例拟合曲线图 'example_dose_response_curve_with_metrics.png'")
else:
    print("没有成功的拟合结果可供绘图")

# 输出拟合质量统计
if not params_df.empty:
    success_rate = params_df['fit_success'].mean() * 100
    avg_r2 = params_df['r_squared'].mean()
    print(f"\n拟合成功率: {success_rate:.1f}%")
    print(f"平均R平方值: {avg_r2:.3f}")
    print("拟合质量分布:")
    print(params_df[['r_squared', 'rmse', 'aic', 'bic']].describe())
```

### 使用lmfit库进行曲线拟合

```python
import pandas as pd
import numpy as np
from lmfit import Model
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error

# 读取数据
file_path = 'data/Hill/Single-dose-response.csv'
df = pd.read_csv(file_path)

# 统计数量
print(f"独立药物数量: {df['drug_name'].nunique()}")
print(f"独立细胞系数量: {df['cell_line'].nunique()}")

# 定义四参数逻辑模型函数
def four_param_logistic(x, top, bottom, ic50, hill_slope):
    return bottom + (top - bottom) / (1 + (x / ic50)**hill_slope)

# 创建lmfit模型
hill_model = Model(four_param_logistic)

# 储存拟合参数和评估指标
fitted_params = []
grouped = df.groupby(['drug_name', 'cell_line'])

for (drug, cell_line), group in grouped:
    concentrations = group['Drug_concentration (µM)'].values
    responses = group['X/X0'].values
    
    # 跳过数据不足的情况
    if len(np.unique(concentrations)) < 3 or len(np.unique(responses)) < 2:
        print(f"数据不足以拟合: {drug} - {cell_line}")
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.nan,
            'bottom': np.nan,
            'ic50': np.nan,
            'hill_slope': np.nan,
            'r_squared': np.nan,
            'adjusted_r_squared': np.nan,
            'rmse': np.nan,
            'aic': np.nan,
            'bic': np.nan,
            'success': False,
            'message': 'Insufficient data points'
        })
        continue
    
    # 设置初始参数和边界
    params = hill_model.make_params()
    params['top'].set(value=max(responses), min=0, max=2.0)
    params['bottom'].set(value=min(responses), min=0, max=2.0)
    params['ic50'].set(value=np.median(concentrations), min=1e-6, max=1e6)
    params['hill_slope'].set(value=1.0, min=0.01, max=10.0)
    
    try:
        # 执行拟合
        result = hill_model.fit(responses, params, x=concentrations)
        
        # 计算预测值
        y_pred = result.eval(x=concentrations)
        
        # 计算各种评估指标
        n = len(responses)
        k = 4  # 4个参数
        
        # R-squared
        r2 = r2_score(responses, y_pred)
        adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
        
        # RMSE
        mse = mean_squared_error(responses, y_pred)
        rmse = np.sqrt(mse)
        
        # AIC和BIC
        rss = np.sum((responses - y_pred)**2)
        aic = n * np.log(rss/n) + 2 * k
        bic = n * np.log(rss/n) + k * np.log(n)
        
        # 保存结果
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': result.best_values['top'],
            'top_stderr': result.params['top'].stderr if result.params['top'].stderr else np.nan,
            'bottom': result.best_values['bottom'],
            'bottom_stderr': result.params['bottom'].stderr if result.params['bottom'].stderr else np.nan,
            'ic50': result.best_values['ic50'],
            'ic50_stderr': result.params['ic50'].stderr if result.params['ic50'].stderr else np.nan,
            'hill_slope': result.best_values['hill_slope'],
            'hill_slope_stderr': result.params['hill_slope'].stderr if result.params['hill_slope'].stderr else np.nan,
            'r_squared': r2,
            'adjusted_r_squared': adj_r2,
            'rmse': rmse,
            'aic': aic,
            'bic': bic,
            'success': result.success,
            'message': result.message,
            'chi_square': result.chisqr,
            'redchi': result.redchi
        })
        
    except Exception as e:
        print(f"拟合失败 {drug}-{cell_line}: {str(e)}")
        # 使用退化模型计算指标
        y_pred = np.full_like(responses, np.mean(responses))
        r2 = r2_score(responses, y_pred)
        adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k - 1)
        mse = mean_squared_error(responses, y_pred)
        rmse = np.sqrt(mse)
        rss = np.sum((responses - y_pred)**2)
        aic = n * np.log(rss/n) + 2 * k
        bic = n * np.log(rss/n) + k * np.log(n)
        
        fitted_params.append({
            'drug_name': drug,
            'cell_line': cell_line,
            'top': np.mean(responses),
            'top_stderr': np.nan,
            'bottom': np.min(responses),
            'bottom_stderr': np.nan,
            'ic50': np.median(concentrations),
            'ic50_stderr': np.nan,
            'hill_slope': 1.0,
            'hill_slope_stderr': np.nan,
            'r_squared': r2,
            'adjusted_r_squared': adj_r2,
            'rmse': rmse,
            'aic': aic,
            'bic': bic,
            'success': False,
            'message': str(e),
            'chi_square': np.nan,
            'redchi': np.nan
        })

# 保存拟合参数
params_df = pd.DataFrame(fitted_params)
output_file = 'data/Hill/fitted_dose_response_parameters_lmfit_enhanced.csv'
params_df.to_csv(output_file, index=False)
print(f"\n拟合参数已保存至 '{output_file}'")
print(params_df.head())

# --- 绘图示例 ---
# 查找第一个成功的拟合结果
success_df = params_df[params_df['success']]
if not success_df.empty:
    example_row = success_df.iloc[0]
    example_drug = example_row['drug_name']
    example_cell_line = example_row['cell_line']
    
    example_data = df[(df['drug_name'] == example_drug) & 
                     (df['cell_line'] == example_cell_line)]
    
    # 生成拟合曲线
    x_fit = np.logspace(np.log10(min(example_data['Drug_concentration (µM)'])), 
                        np.log10(max(example_data['Drug_concentration (µM)'])), 100)
    y_fit = four_param_logistic(
        x_fit,
        example_row['top'],
        example_row['bottom'],
        example_row['ic50'],
        example_row['hill_slope']
    )
    
    # 绘图设置
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus'] = False
    
    plt.figure(figsize=(10, 7))
    plt.scatter(example_data['Drug_concentration (µM)'], example_data['X/X0'], 
               label='原始数据')
    plt.plot(x_fit, y_fit, 'r-', label='4PL 拟合曲线')
    plt.xscale('log')
    plt.xlabel('药物浓度 (µM) (对数坐标)')
    plt.ylabel('相对抑制率 (X/X0)')
    
    # 在标题中添加更多指标
    plt.title(
        f'药物: {example_drug} | 细胞系: {example_cell_line}\n'
        f'IC50: {example_row["ic50"]:.3f} µM (±{example_row["ic50_stderr"]:.3f}) | '
        f'Hill斜率: {example_row["hill_slope"]:.2f}\n'
        f'R²: {example_row["r_squared"]:.3f} | 调整R²: {example_row["adjusted_r_squared"]:.3f}\n'
        f'RMSE: {example_row["rmse"]:.3f} | AIC: {example_row["aic"]:.1f}'
    )
    
    plt.legend()
    plt.grid(True)
    plt.savefig('example_dose_response_curve_lmfit_enhanced.png')
    print("\n已保存示例拟合曲线图 'example_dose_response_curve_lmfit_enhanced.png'")
else:
    print("没有成功的拟合结果可供绘图")

# 输出整体拟合质量统计
if not params_df.empty:
    print("\n===== 整体拟合质量统计 =====")
    success_rate = params_df['success'].mean() * 100
    print(f"拟合成功率: {success_rate:.1f}%")
    
    successful_fits = params_df[params_df['success']]
    if not successful_fits.empty:
        print("\n成功拟合的统计信息:")
        print(successful_fits[['r_squared', 'adjusted_r_squared', 'rmse', 'aic', 'bic']].describe())
    
    failed_fits = params_df[~params_df['success']]
    if not failed_fits.empty:
        print("\n失败拟合的原因统计:")
        print(failed_fits['message'].value_counts())
```

### 删除拟合质量不高的数据r_squre<0.7

```python
import pandas as pd

def filter_r_squared(input_filename, output_filename, threshold=0.7):
    """
    Reads a CSV file, filters it based on an r_squared threshold,
    and saves the result to a new CSV file.

    Args:
        input_filename (str): The name of the input CSV file.
        output_filename (str): The name of the output CSV file.
        threshold (float, optional): The r_squared threshold. Defaults to 0.7.
    """
    try:
        # Read the CSV file into a pandas DataFrame
        df = pd.read_csv(input_filename)

        # Check if 'r_squared' column exists
        if 'r_squared' not in df.columns:
            print(f"Error: Column 'r_squared' not found in {input_filename}")
            return

        # Get the number of rows before filtering
        initial_rows = len(df)
        print(f"Initial number of rows: {initial_rows}")

        # Filter the DataFrame
        df_filtered = df[df['r_squared'] >= threshold]

        # Get the number of rows after filtering
        filtered_rows = len(df_filtered)
        print(f"Number of rows after filtering: {filtered_rows}")

        # Calculate the number of removed rows
        removed_rows = initial_rows - filtered_rows
        print(f"Number of rows removed: {removed_rows}")

        # Save the filtered DataFrame to a new CSV file
        df_filtered.to_csv(output_filename, index=False)

        print(f"\nFiltered data saved to '{output_filename}'")
        print("\nFirst 5 rows of the filtered data:")
        print(df_filtered.head())

    except FileNotFoundError:
        print(f"Error: The file '{input_filename}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- Main execution ---
if __name__ == "__main__":
    # Define the input and output filenames
    input_csv = 'data/Hill/evaluated_fitted_parameters.csv'
    output_csv = 'remove0.7_doseresponse.csv'

    # Call the function to filter the data
    filter_r_squared(input_csv, output_csv)
```
